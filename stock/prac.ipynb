```python
# Pull daily close prices for TSLA over last 7 days

import yfinance as yf
import matplotlib.pyplot as plt

recent_data = yf.download("TSLA", period="7D", auto_adjust=True)
recent_data['Close'].plot(label="TSLA")
plt.legend(title="Ticker")
plt.title("Tesla Closing Prices Across Last 7 Days")
plt.show()
```

```python
from dotenv import load_dotenv
import os

load_dotenv(dotenv_path="tiingo.env")
api_key = os.getenv("tiingo_key")

print(api_key)
```

```python
# GET TSLA price data from Tiingo
import requests

ticker = "TSLA"
url = f"https://api.tiingo.com/tiingo/daily/{ticker}/prices?startDate=2024-05-01&endDate=2024-06-13"

headers = {
    'Content-Type': 'application/json',
    'Authorization': f'Token {api_key}'
}

response = requests.get(url, headers=headers)
price_data = response.json()

import pandas as pd
df = pd.DataFrame(price_data)
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
df.head()
```
